# DataFrame Joins

* So now we have two DataFrames which have some common names, 
and we want to create a single DataFrame with both the Age and Skill information.

* Notice that some names are different in df1 and df2

* There are four ways in which we can combine the two DataFrames:

  * Inner Join : Keep only those names which are common in both DataFrames
  * Left Join : Keep all those names which are there in df1
  * Right Join : Keep all those names which are there in df2
  * Outer Join : Keep all names which are in df1 or df2

```python
# DataFrame is a class in pandas which helps user to create their own dataframes.
df1 = pd.DataFrame()        #Creating empty dataframe.

# Putting data into data frame.
df1["Name"] = ["Rahul", "Ramesh", "Rakesh", "Roshan", "Ricky", "Rambo"]
df1["Age"] = [23, 25, 21, 28, 27, 28]
df1
```

```python
# create another dataframe.
df2 = pd.DataFrame()

df2["Name"] = ["Rahul", "Rashmi", "Rakesh", "Roshan", "Ramona", "Rambo"]
df2["Skill"] = ["Music", "Dance", "Painting", "Music", "Math", "Programming"]
df2
```

**Inner Join**

```python
# Merge function is used to merge two dataframes into one.
df_inner = pd.merge(df1, df2, on='Name', how='inner')
```

```python
# Notice that only the common names present in both df1 and df2 have been retained
df_inner
```

**Left Join**

```python
df_left = pd.merge(df1, df2, on='Name', how='left')
```

```python
# Notice that the names from df1 that are missing in df2 now have NaN in the column
df_left
```

**Right Join**

```python
df_right = pd.merge(df1,df2, on='Name', how='right')
```

```python
# Notice that the names from df2 that are missing in df1 now have NaN in the column
df_right
```

**Outer Join**

```python
df_outer = pd.merge(df1, df2, on='Name', how='outer')
```

```python
# Notice that missing names in either column have NaN in the column
df_outer
```
